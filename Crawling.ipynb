```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import numpy as np
import time
```

```python
#Selenium driver
driver = webdriver.Chrome()
```

```python
#The columns of my future dataframe
MovieName = [] 
MovieGenre = [] #multi label
MovieRate = []
MovieDescription = [] 
MovieMetascore = [] 
MovieDateOfRelease = [] 
MovieLength = [] 
MovieRestriction = []
MovieGross = [] 
MovieVotes = [] 
```

```python
#Genres of all movies in IMDB website
Genres = ['Action', 'Adventure','Animation','Biography','Comedy','Crime','Documentary','Drama',
'Family','Fantasy','Film-Noir','History','Horror','Music','Musical','Mystery','Romance','Sci-Fi',
'Short','Sport','Superhero','Thriller','War','Western']
```

```python
#Using 2 urls in order to concat them together in the next block
Url1 = 'https://www.imdb.com/search/title/?genres='
Url2 = '&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=5aab685f-35eb-40f3-95f7-c53f09d542c3&pf_rd_r=X2PECMFHNTZA7FYDP1HP&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1'
```

```python
for genre in Genres:
    Url = Url1 + genre + Url2 #The concat of 2 urls + the genre
    print(Url)
    print(genre)
    page = driver.get(Url)
    next_flag = 1
    while next_flag == 1: #This loop is for getting to next page of the website
        time.sleep(3)
        #From here is the crawling code. Filling my columns with data
        try:
            MovieDivs = driver.find_elements(By.CLASS_NAME, 'lister-item')
        except:
            driver.refresh()
            continue
        for movie in MovieDivs:
            num_of_movie_ser = ''
            year = ''
            try:
                header = movie.find_element(By.CLASS_NAME,"lister-item-header")
                txt = header.find_element(By.CLASS_NAME,'lister-item-year').text
                try:
                    year = txt.split()[1].strip('()')
                    num_of_movie_ser = txt.split(' ')[0].strip('()')
                except:
                    year = txt.split()[0].strip('()')
            except:
                year = np.nan
            MovieDateOfRelease.append(year)
            try:
                header = movie.find_element(By.CLASS_NAME,"lister-item-header")
                name = header.find_element(By.TAG_NAME,'a').text
                if num_of_movie_ser != '':
                    name = "{0} {1}".format(name,num_of_movie_ser)
            except:
                name = np.nan
            MovieName.append(name)  
            try:
                Restriction = movie.find_element(By.CLASS_NAME, 'certificate').text
            except:
                Restriction = np.nan
            MovieRestriction.append(Restriction)
            try:
                gen = movie.find_element(By.CLASS_NAME, 'genre').text
            except:
                gen = np.nan
            MovieGenre.append(gen)
            try:
                length = movie.find_element(By.CLASS_NAME, 'runtime').text.split()[0]
            except:
                length = np.nan
            MovieLength.append(length)
            try:
                ratedivs = movie.find_element(By.CLASS_NAME, 'inline-block')
                rate = ratedivs.find_element(By.TAG_NAME, 'strong').text
            except:
                rate = np.nan
            MovieRate.append(rate)
            try:
                metadiv = movie.find_element(By.CLASS_NAME, 'ratings-metascore')
                metascore = metadiv.find_element(By.CLASS_NAME, 'metascore').text
            except:
                metascore = np.nan
            MovieMetascore.append(metascore)
            try:
                desc = movie.find_elements(By.CLASS_NAME, 'text-muted')[2].text
            except:
                desc = np.nan
            MovieDescription.append(desc)
            try:
                p = movie.find_element(By.CLASS_NAME, 'sort-num_votes-visible')
                votes = p.find_elements(By.NAME, 'nv')[0].get_attribute('data-value')
            except:
                votes = np.nan
            MovieVotes.append(votes)
            try:
                gross = p.find_elements(By.NAME, 'nv')[1].get_attribute('data-value')
            except:
                gross = np.nan
            MovieGross.append(gross)         
        try:
            title_page = driver.find_element(By.CLASS_NAME, 'desc')
            next_page = title_page.find_element(By.CLASS_NAME, 'next-page').get_attribute('href')
            driver.get(next_page)
        except:
            next_flag = 0
```

```python
dictionary = {'Movie_Name': MovieName, 'Movie_Description':MovieDescription, 
              'Movie_Date_Of_Release': MovieDateOfRelease, 
              'Movie_Genres': MovieGenre, 'Movie_Rate': MovieRate, 
              'Movie_Metascore': MovieMetascore, 
              'Movie_Length': MovieLength, 'Movie_Votes': MovieVotes, 
              'Movie_Restriction': MovieRestriction, 
              'Movie_Gross': MovieGross}
```

```python
df = pd.DataFrame(dictionary)
```

```python
df.to_csv('IMDB_Data.csv')
```
