```python
import pandas as pd
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import spacy
from spacy import displacy
```

```python
#Uploading the dataframe and deleting the unnamed:0 column from it
df = pd.read_csv('Complete.csv')
```

```python
#Split into 2 different datasets in order to work only with description and tokenized data
df_description = df['Movie_Description'].to_frame(name = None)
df_labels = df.iloc[:,1:]
```

```python
#Every words I need to remove like names locations and etc..
nlp = spacy.load('en_core_web_sm')
df_description['ents'] = df_description['Movie_Description'].apply(lambda sen: nlp(sen))
```

```python
#I wanna insert every entity to the dataframe. Took the entities from stack overflow
Gpe = []
Person = []
Loc = []
Org = []
Fac = []
Norp = []
Date = []
Work_Of_Art = []
Event = []
Cardinal = []
Money = []
Quantity = []
Ordinal = []
Product = []
Time = []
Language = []
Updated = []
```

```python
#Counting every entity appereance in words and appending to my arrays for df
for i,row in df_description.iterrows():
    CountGpe = 0
    CountPerson = 0
    CountLoc = 0
    CountOrg = 0
    CountFac = 0
    CountNorp = 0
    CountDate = 0
    CountWorkOfArt = 0
    CountEvent = 0
    CountCardinal = 0
    CountQuantity = 0
    CountMoney = 0
    CountProduct = 0
    CountTime = 0
    CountOrdinal = 0
    CountLanguage = 0
    desc = row['Movie_Description']
    for word in row['ents'].ents:
        if word.label_ == 'GPE':
            CountGpe += 1
        elif word.label_ == 'PERSON':
            CountPerson += 1
            desc = desc.replace(str(word), '')
        elif word.label_ == 'LOC':
            CountLoc += 1
        elif word.label_ == 'ORG':
            CountOrg += 1
            desc = desc.replace(str(word), '')
        elif word.label_ == 'FAC':
            CountFac += 1
        elif word.label_ == 'NORP':
            CountNorp += 1
        elif word.label_ == 'DATE':
            CountDate += 1
        elif word.label_ == 'WORK_OF_ART':
            CountWorkOfArt += 1
        elif word.label_ == 'EVENT':
            CountEvent += 1
        elif word.label_ == 'CARDINAL':
            CountCardinal += 1
            desc = desc.replace(str(word), '')
        elif word.label_ == 'MONEY':
            CountMoney += 1
            desc = desc.replace(str(word), '')
        elif word.label_ == 'QUANTITY':
            CountQuantity += 1
            desc = desc.replace(str(word), '')
        elif word.label_ == 'PRODUCT':
            CountProduct += 1
        elif word.label_ == 'TIME':
            CountTime += 1
            desc = desc.replace(str(word), '')
        elif word.label_ == 'ORDINAL':
            CountOrdinal += 1
            desc = desc.replace(str(word), '')
        elif word.label_ == 'LANGUAGE':
            CountLanguage += 1
    Gpe.append(CountGpe)
    Person.append(CountPerson)
    Loc.append(CountLoc)
    Org.append(CountOrg)
    Fac.append(CountFac)
    Norp.append(CountNorp)
    Date.append(CountDate)
    Work_Of_Art.append(CountWorkOfArt)
    Event.append(CountEvent)
    Cardinal.append(CountCardinal)
    Money.append(CountMoney)
    Quantity.append(CountQuantity)
    Product.append(CountProduct)    
    Time.append(CountTime)
    Ordinal.append(CountOrdinal)
    Language.append(CountLanguage)
    Updated.append(desc)
```

```python
df_description['Filtered_Description'] = Updated
df_description['GPE'] = Gpe 
df_description['PERSON'] = Person
df_description['LOC'] = Loc 
df_description['ORG'] = Org 
df_description['FAC'] = Fac 
df_description['NORP'] = Norp
df_description['DATE'] = Date 
df_description['WORK_OF_ART'] = Work_Of_Art 
df_description['EVENT'] = Event 
df_description['CARDINAL'] = Cardinal
df_description['TIME'] = Time
df_description['QUANTITY'] = Quantity 
df_description['MONEY'] = Money 
df_description['ORDINAL'] = Ordinal 
df_description['PRODUCT'] = Product
df_description['LANGUAGE'] = Language
```

```python
#Converting every word in dataframe to lower case
df_description['Movie_Description'] = df_description['Movie_Description'].str.lower()
```

```python
#Tokenization and removing punctuations from the dataframe
df_description['Tokenized'] = df_description['Movie_Description'].apply(lambda row : 
list(filter(lambda token: token not in string.punctuation, word_tokenize(row))))
```

```python
#Removing all stop words from dataframe
StopWords = stopwords.words('english')
df_description['Tokenized'] = df_description['Tokenized'].apply(
lambda words:[x for x in words if x not in StopWords])
```

```python
#Counting how many words I have in movie description
df_description['Word Count'] = df_description['Movie_Description'].apply(
lambda x: len((x).split()))
```

```python
#lemmatizer function in order to stem every word
lemmatizer = WordNetLemmatizer()
```

```python
#Stemming every word in the dataframe
df_description['Stemmed'] = df_description['Tokenized'].apply(
lambda x:([lemmatizer.lemmatize(word) for word in x]))
```
