```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('IMDB_DATA1.csv')
df2 = pd.read_csv('IMDB_DATA2.csv')
```

```python
del df1['Unnamed: 0']
del df2['Unnamed: 0']
```

```python
frames = [df1, df2]
resultdf = df1.append(df2)
```

```python
display(resultdf)
```

```python
resultdf.drop_duplicates(inplace = True, subset = ['Movie_Name'])
```

```python
resultdf.info()
```

```python
resultdf.drop(['Movie_Gross', 'Movie_Metascore'], axis=1, inplace = True)
resultdf = resultdf.dropna()
resultdf.reset_index(inplace = True, drop = True)
```

```python
#Reduced the number of genres that i wanna check. the prev list was too long
genres_to_clean = ['Action', 'Sci-Fi', 'Comedy', 'Crime', 'Western','Family',
          'Drama', 'Horror', 'Romance', 'Thriller','Adventure']
```

```python
#Apply does the function again and again on all rows
#lambda is an anonymous func like: def do(row):  <-- the same as the code here below
#np.isin must compare to lists so i checked if genres contain row.split(',')
new = resultdf['Movie_Genres'].apply(lambda row : np.isin(genres_to_clean, row.split(', ')).astype(int))
```

```python
#In order to build a dataframe it must be list of list so i converted the new to list
#Created new dataframe of genres only
#After this i concat in order to combine everything together
resdf = pd.DataFrame(new.to_list(), columns = genres_to_clean)
```

```python
#concat of 2 dataframes to one dataframe
result = pd.concat([resultdf,resdf],axis = 1)
```

```python
result['Movie_Restriction'].value_counts()
```

```python
#Dropping the columns that i dont need for my learning
result.drop(['Movie_Name',
            'Movie_Date_Of_Release','Movie_Rate','Movie_Length','Movie_Votes','Movie_Restriction'
            ,'Movie_Genres'], axis=1, inplace = True)
```

```python
result.to_csv('complete.csv')
```
